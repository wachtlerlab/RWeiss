```python
from scipy import fftpack

fs = 512
w = 2*pi*8.
x = arange(0,1,1./fs)

f = sin(w*x)

plot(x, f)

F = fftpack.fft(f)/fs
freq = fftpack.fftfreq(fs, 1./fs)

print abs(F[:20])


figure()
plot(freq[:20], abs(F)[:20], 'o')
```
