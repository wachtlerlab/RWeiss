```python
from scipy import fftpack

fs = 512
w = 2*pi*4.
x = arange(0,1,1./fs)

f = sin(w*x)

a = plot(x, f)
plt.xlabel('$t$ in Sekunden')
plt.ylabel('$sin(8\pi x)$')

F = fftpack.fft(f)/fs
freq = fftpack.fftfreq(fs, 1./fs)

print abs(F[:20])


figure()
afft = plot(freq[:20], abs(F)[:20], 'o')
plt.xlabel('Frequenz in $Hz$')

plt.close('all')
fig, (a, afft) = plt.subplots(nrows=1, ncols=2)
a.plot(x, f)
a.set_xlabel('$t$ in Sekunden')
a.set_ylabel('$sin(8\pi x)$')
afft.plot(freq[:20], abs(F)[:20], 'o')
afft.set_xlabel('Frequenz in $Hz$')
plt.tight_layout()
```
