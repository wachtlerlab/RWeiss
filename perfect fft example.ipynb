```python
from scipy import fftpack

fs = 512
w = 2*pi*4.
x = arange(-0.06,1.44,1./fs)

f = sin(w*x) + 0.5*sin(2*w*x+0.85*pi) + sin(.5*w*x)
plt.figure(figsize=(10,4))
ax = plt.subplot(111)

a = plot(x, f, linewidth=3)
a1 = plot((0.09,0.09), (0,2.5), linewidth=2, color='g')
a2 = plot((0.59,0.59), (0,2.5), linewidth=2, color='g')
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel('t in Sekunden', fontsize=12)
plt.ylabel('f(t)', fontsize=12)
plt.grid()
plt.yticks([0])
plt.xticks([0, 1])

#plt.fill_between(x, 0, f, alpha=.3)



F = fftpack.fft(f)/fs
freq = fftpack.fftfreq(fs, 1./fs)



figure()
afft = plot(freq[:20], abs(F)[:20], 'or')
plt.xlabel('Frequenz in Hz', fontsize=12, fontname='serif')
plt.grid()
plt.ylim(-.1, .6)

plt.yticks([0, .25, .5])

'''
plt.close('all')
fig, (a, afft) = plt.subplots(nrows=1, ncols=2)
a.plot(x, f)
a.set_xlabel('$t$ in Sekunden')
a.set_ylabel('$sin(8\pi x)$')
afft.plot(freq[:20], abs(F)[:20], 'o')
afft.set_xlabel('Frequenz in $Hz$')
plt.tight_layout()
'''
```
