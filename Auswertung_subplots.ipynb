```python
import glob
import csv
from operator import add

#Dateiliste erstellen
files = []
for filename in glob.glob('/home/weiss/data/T*.csv'):
    files.append(filename)
    
#read files
seen = 0
trials = 0
catchseen = 0
catchtrials = 0
deltas = []
freqs = []
for filename in files:
    data = csv.reader(open(filename, "rb"), delimiter=',')
    data.next()
    data.next()
    data.next()
    for row in data:
        freqdiff = abs(int(row[0][0:-2])-int(row[1][0:-2]))
        if row[3] < row[4]:
            if row[0] != row[1]:
                trials += 1
                if row[5] == 'y':
                    seen += 1
                    deltas.append((freqdiff,1))
                    freqs.append((int(row[0][0:-2]),int(row[1][0:-2]),1))
                else:
                    deltas.append((freqdiff,0))
                    freqs.append((int(row[0][0:-2]),int(row[1][0:-2]),0))
            if row[0] == row[1]:
                catchtrials += 1
                if row[5] == 'n':
                    catchseen += 1
                                        
d_delta = {}
for freq, count in deltas:
    if freq not in d_delta:
        d_delta[freq] = [0,0]
    d_delta[freq][0] += 1
    if count:
        d_delta[freq][1] += 1
        
d_freqs = {}
for freq1, freq2, count in freqs:
    if (freq1, freq2) not in d_freqs:
        d_freqs[(freq1, freq2)] = [0,0]
    d_freqs[(freq1, freq2)][0] += 1
    if count:
        d_freqs[(freq1, freq2)][1] += 1
        

d_all = d_freqs.copy()
d_all['overall\ncorrectness'] = (1, round(float(seen)/trials, 3))
d_all['false\npositives'] = (1, 1-round(float(catchseen)/catchtrials, 3))

print 'seeing-prob:', round(100.*seen/trials, 3), '%'
print 'catchtrial accuracy:', round(100.*catchseen/catchtrials), '%'


f, p = plt.subplots(3, figsize=(15,12))

p1 = p[0].bar(range(len(d_delta)), [d_delta[key][0] for key in sorted(d_delta)], align='edge', width=0.1, color='c')
p2 = p[0].bar(range(len(d_delta)), [d_delta[key][1] for key in sorted(d_delta)], align='edge', width=-0.1, color='b')

p[0].set_xticks(range(len(d_delta)))
p[0].set_xticklabels(sorted(d_delta))
p[0].set_xlabel('$\Delta f$ in $Hz$')
p[0].set_ylabel('Anzahl')
p[0].set_title('Richtige Antworten pro $\Delta f$')
p[0].legend((p2, p1), ('gesehen', 'gezeigt'))
p[0].grid()
p[0].set_ylim(0,70)

p[1].bar(range(len(d_all)), [float(d_all[key][1])/d_all[key][0] for key in sorted(d_all)], align='center', width=.2, color='m')
p[1].set_xticks(range(len(d_all)))
p[1].set_xticklabels(sorted(d_all))
p[1].set_ylim(0, 1.1)
p[1].set_xlim(-1, 14)
p[1].grid()
p[1].set_title('Probability of Seeing')
p[1].set_xlabel('Frequenzpaar')
p[1].set_ylabel('P(seeing)')


contrastsum = []
for key in sorted(d_freqs):
    if key[0]<key[1]:
        contrastsum.append(map(add, d_freqs[key], d_freqs[tuple(reversed(key))]))
                
contrasts = [float(fr[1])/fr[0]-1 for fr in sorted(d_freqs) if fr[0]<fr[1]]
p[2].bar(range(len(contrasts)), [(float(item[1])/item[0]) for item in contrastsum], align='center', width=.1, color='y')
p[2].set_xticks(range(len(contrasts)))
p[2].set_xticklabels(contrasts)
p[2].grid()
p[2].set_ylim(0, 1.1)
p[2].set_xlabel('contrast')
p[2].set_ylabel('P(seeing)')

f.tight_layout()
```
